```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# mlxtend 패키지 apriori, association_rules 모듈 호출
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv('/content/retail_dataset.csv')
df.head(10)
```

```python
df.info()
```

# 전처리 작업

```python
# 장바구니 데이터 고유항목 구분 출력
items = set()  # 빈 집합 set 객체를 생성
for col in df:
    items.update(df[col].unique())

# 장바구니 목록 값(텍스트)을 수치로 표현-각 항목당 매칭될 경우 1로, 아니면 0으로 표시(one-hot encoding)
itemset = set(items)
encoding = []
for index, row in df.iterrows():
    rowset = set(row)
    labels = {}
    dismatching = list(itemset - rowset)
    matching = list(itemset.intersection(rowset))  # intersection : 교집합 구하기
    for i in dismatching:
        labels[i] = 0
    for j in matching:
        labels[j] = 1
    encoding.append(labels)

result = pd.DataFrame(encoding)
result
```

```python
#NaN열 제외시키기
result=result.drop(result.columns[1],axis=1)
result
```

```python
df.info()
```

# 데이터 분석

```python
#apriori함수 적용(빈번하게 등장하는 상품 집합 찾기)
freq_items=apriori(result,min_support=0.2,use_colnames=True)  #지지도0.2이상만 보여줌
freq_items
```

```python
rules=association_rules(freq_items,metric='confidence',min_threshold=0.6)
rules
#두번째줄(Cheese → Milk)
#antecedent suppor: 전체중 치즈가 등장한 비율 0.501587
#Consequent Support =전체중 Milk가 등장한 비율  0.501587
#Support = 전체 거래 중 Cheese와 Milk가 함께 등장한 비율 0.304762 (30.48%)
#Confidence =Cheese가 등장한 거래 중 Milk도 등장한 비율  0.607595 (60.76%)
```
