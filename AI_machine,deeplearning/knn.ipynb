```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df.head(10)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler   #KNN에선 min-max정규화 진행함
scaler=MinMaxScaler()
df['sepal_length']=scaler.fit_transform(df[['sepal_length']])
df['sepal_width']=scaler.fit_transform(df[['sepal_width']])
df['petal_length']=scaler.fit_transform(df[['petal_length']])
df['petal_width']=scaler.fit_transform(df[['petal_width']])
df.head(10)
```

```python
X=df[['sepal_length','sepal_width','petal_length','petal_width']]  #독립변수
y=df['species']  #종속변수
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,pred)
print(acc)
```

```python
mat=confusion_matrix(y_test,pred)  #성능평가
print(mat)

# 예측 0	예측 1	예측 2
# 실제 0	14	0	0
# 실제 1	0	12	2
# 실제 2	0	2	15

# 클래스 0: 14개를 정확하게 맞춤
# 클래스 1: 12개를 정확하게 맞춤
# 클래스 2: 15개를 정확하게 맞춤

# 오분류된 경우 (비대각선 값)
# 클래스 1의 샘플 2개가 클래스 2로 잘못 예측됨
# 클래스 2의 샘플 2개가 클래스 1로 잘못 예측됨
```

```python
rep=classification_report(y_test,pred)
print(rep)
```

# SVM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df
```

```python
df.info()
```

```python
d_mean=df['Age'].mean()
df['Age'].fillna(d_mean,inplace=True)
```

```python
d_mode=df['Embarked'].mode()
df['Embarked'].fillna(d_mode[0],inplace=True)
```

```python
df['FamilySize']=df['SibSp']+df['Parch']
```

```python
onehot_sex=pd.get_dummies(df['Sex'])  #원-핫 인코딩(크기기준X)
df=pd.concat([df,onehot_sex],axis=1)

onehot_embarked=pd.get_dummies(df['Embarked'])
df=pd.concat([df,onehot_embarked],axis=1)
```

```python
df.head(10)
```

```python
X=df[['Pclass','Age','Fare','FamilySize','female','male','C','Q','S']]  #독립변수
y=df['Survived']  #종속변수
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=svm.SVC(kernel='rbf')
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
mat=confusion_matrix(y_test,pred)  #성능평가
print(mat)
#         예측 0  예측 1
# 실제 0	 162     14
# 실제 1	  67      25
```

```python
rpt=classification_report(y_test,pred)
print(rpt)
```
