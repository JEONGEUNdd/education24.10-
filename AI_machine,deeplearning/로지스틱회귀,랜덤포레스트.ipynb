```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
scaler=MinMaxScaler()
df[["sepal_length"]]=scaler.fit_transform(df[["sepal_length"]])
df[["sepal_width"]]=scaler.fit_transform(df[["sepal_width"]])
df[["petal_length"]]=scaler.fit_transform(df[["petal_length"]])
df[["petal_width"]]=scaler.fit_transform(df[["petal_width"]])
df.head()
```

```python
df.describe()
```

```python
X=df[["sepal_length","sepal_width","petal_length","petal_width"]]
y=df["species"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
mat=confusion_matrix(y_test,pred)
print(mat)
```

```python
rpt=classification_report(y_test,pred)
print(rpt)
```

# 랜덤포레스트

```python
import numpy as np
import pandas as pd
import sklearn

#로지스틱회귀분류모델
from sklearn.ensemble import RandomForestClassifier

#학습 및 테스트 데이터셋 분리 위한 패키지
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head(10)
```

```python
#Age컬럼의 결측값을 평균으로 대치
d_mean=df['Age'].mean()
df['Age'].fillna(d_mean,inplace=True)

#Embarked(출발항구코드)컬럼의 결측값을 최빈값으로 대치
d_mode=df['Embarked'].mode()
df['Embarked'].fillna(d_mode[0],inplace=True)
```

```python
#1,0의 레이블로 인코딩
from sklearn.preprocessing import LabelEncoder
df['Sex']=LabelEncoder().fit_transform(df['Sex'])
df['Embarked']=LabelEncoder().fit_transform(df['Embarked'])
```

```python
df['FamilySize']=df['SibSp']+df['Parch']
```

```python
X=df[["Pclass",'Sex','Age','Embarked','FamilySize']]
y=df['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#RandomForestClassification객체생성
rf=RandomForestClassifier(n_estimators=50,max_depth=3,random_state=11)
rf.fit(X_train,y_train)
```

```python
pred=rf.predict(X_test) #학습 완료된 객체에서 예측 수행
```

```python
from sklearn.metrics import pair_confusion_matrix
mat=confusion_matrix(y_test,pred)
print(mat)
```

```python
#모델 성능 평가
from sklearn.metrics import classification_report
rpt=classification_report(y_test,pred)
print(rpt)   #svm에서 0.7정확도였는데 랜덤포레스트에서 0.88로 상승
```
