```python
# 1. 티스토리 블로그 크롤링
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

def tistory_blog():
    url = "https://www.tistory.com/?category=living"
   
    # Chrome 옵션 설정
    options = Options()
    options.add_argument('--headless')  # 브라우저가 보이지 않게 실행
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    # Chrome 드라이버 경로 설정
    service = Service('chromedriver.exe')  # chromedriver 경로 지정
    driver = webdriver.Chrome(service=service, options=options)
   
    # 웹 페이지 열기
    driver.get(url)
   
    # 페이지 로딩 대기
    time.sleep(5)  # 필요에 따라 대기 시간 조정
   
    posts = []
   
    # 블로그 포스트 정보 가져오기
    items = driver.find_elements(By.CSS_SELECTOR, 'div.item_group')
    for item in items:
        title = item.find_element(By.CSS_SELECTOR, '.tit_cont').text.strip()
        desc = item.find_element(By.CSS_SELECTOR, '.wrap_desc').text.strip()
        author = item.find_element(By.CSS_SELECTOR, '.link_blog .txt_g').text.strip()
       
        posts.append({
            'title': title,
            'description': desc,
            'author': author
        })
   
    # HTML 생성
    html_content = f'''
    <html>
    <head>
        <style>
            .blog-container {{
                max-width: 900px;
                margin: 20px auto;
                font-family: 'Noto Sans KR', sans-serif;
            }}
            .blog-post {{
                background: #fff;
                padding: 20px;
                margin: 15px 0;
                box-shadow: 0 2px 5px rgba(0,0,0,0.1);
                border-radius: 8px;
            }}
            .blog-title {{
                font-size: 20px;
                color: #333;
                margin-bottom: 10px;
            }}
            .blog-desc {{
                color: #666;
                line-height: 1.6;
            }}
            .blog-author {{
                color: #888;
                font-size: 14px;
                margin-top: 10px;
            }}
        </style>
    </head>
    <body>
        <div class="blog-container">
    '''
   
    for post in posts:
        html_content += f'''
            <div class="blog-post">
                <div class="blog-title">{post['title']}</div>
                <div class="blog-desc">{post['description']}</div>
                <div class="blog-author">작성자: {post['author']}</div>
            </div>
        '''
   
    html_content += '''
        </div>
    </body>
    </html>
    '''
   
    # HTML 파일 저장
    with open('tistory_posts.html', 'w', encoding='utf-8') as f:
        f.write(html_content)

    driver.quit()  # 드라이버 종료

# 크롤링 함수 호출
tistory_blog()
```

```python
#뉴스 추출(html)
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import os
import requests
import httpx
import ssl

ssl_context=ssl.create_default_context()
ssl_context.options |= ssl.OP_LEGACY_SERVER_CONNECT

def news():
    url = "https://www.yna.co.kr/theme/hotnews-history"
    if not os.path.exists('images'):
        os.makedirs('images')
        
    # Chrome 옵션 설정
    options = Options()
    options.add_argument('--headless')  # 브라우저가 보이지 않게 실행
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    # Chrome 드라이버 경로 설정
    service = Service('chromedriver.exe')  # chromedriver 경로 지정
    driver = webdriver.Chrome(service=service, options=options)
   
    # 웹 페이지 열기
    driver.get(url)
   
    # 페이지 로딩 대기
    time.sleep(5)  # 필요에 따라 대기 시간 조정
   
    posts = []
    nn = 1  
    # 블로그 포스트 정보 가져오기
    items = driver.find_elements(By.CSS_SELECTOR, '.item-box01')  #전체 아니고 각 뉴스박스 선택
    for item in items:
        title = item.find_element(By.CSS_SELECTOR, '.tit-news').text.strip()
        desc = item.find_element(By.CSS_SELECTOR, '.lead').text.strip()
        times = item.find_element(By.CSS_SELECTOR, '.txt-time').text.strip()
        img_element=item.find_element(By.CSS_SELECTOR, 'img')
        img_url=img_element.get_attribute('src')

        if img_url:
            response=httpx.get(img_url,verify=ssl_context)
            #response=requests.get(img_url)
            filename=f'images/article{nn}.jpf'
            if response.status_code==200:
                with open(filename,'wb') as f:
                    f.write(response.content)
                nn += 1 
        else:
            filename=None

            
        posts.append({
            'title': title,
            'description': desc,
            'times': times
        })


     # HTML 생성
    html_content = f'''
    <html>
    <head>
        <style>
            .blog-container {{
                max-width: 900px;
                margin: 20px auto;
                font-family: 'Noto Sans KR', sans-serif;
            }}
            .blog-post {{
                background: #fff;
                padding: 20px;
                margin: 15px 0;
                box-shadow: 0 2px 5px rgba(0,0,0,0.1);
                border-radius: 8px;
            }}
            .blog-title {{
                font-size: 20px;
                color: #333;
                margin-bottom: 10px;
            }}
            .blog-desc {{
                color: #666;
                line-height: 1.6;
            }}
            .blog-times {{
                color: #888;
                font-size: 14px;
                margin-top: 10px;
            }}
        </style>
    </head>
    <body>
        <div class="blog-container">
    '''
   
    for post in posts:
        html_content += f'''
            <div class="blog-post">
                <div class="blog-title">{post['title']}</div>
                <div class="blog-desc">{post['description']}</div>
                <div class="blog-times">작성일자: {post['times']}</div>
                <img src="{filename}" alt="{post['title']}" style="width:100px; height:100px">
            </div>
        '''
   
    html_content += '''
        </div>
    </body>
    </html>
    '''
    
    # HTML 파일 저장
    with open('news1.html', 'w', encoding='utf-8') as f:
        f.write(html_content)

    driver.quit()  
news()
```

```python
#selenium이용 json
#네이버정보 가져오기
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

def naverblog():
    url = "https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0"

    # Chrome 옵션 설정
    options = Options()
    options.add_argument('--headless')  # 브라우저가 보이지 않게 실행
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36')

    # Chrome 드라이버 경로 설정
    service = Service('chromedriver.exe')  # chromedriver 경로 지정
    driver = webdriver.Chrome(service=service, options=options)
   
    # 웹 페이지 열기
    driver.get(url)

    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);") 
    #selenium는 스크립트를 불러올 수 있음 그래서 BeautifulSoup에선 35개 불러왔는데 
    #selenium에선 66개 불러옴

    
    # 페이지 로딩 대기
    time.sleep(5)  # 필요에 따라 대기 시간 조정
   
    posts = []
   
    # 블로그 포스트 정보 가져오기
    items = driver.find_elements(By.CSS_SELECTOR, '.view_wrap')
    
    for item in items:
        title = item.find_element(By.CSS_SELECTOR, '.title_area').text.strip()
        desc = item.find_element(By.CSS_SELECTOR, '.dsc_area').text.strip()
        link = item.find_element(By.CSS_SELECTOR, '.title_area a').get_attribute('href')
        #링크 가져오는 방식이 BeautifulSoup에서랑 다름(get_attribute)
        posts.append({
            'title': title,
            'description': desc,
            'link': link
        })
    
    #json 파일 저장
    with open('naverblog.json','w',encoding='utf-8') as f:
        json.dump(posts,f,ensure_ascii=False)

    driver.quit()  # 드라이버 종료
naverblog()
```
