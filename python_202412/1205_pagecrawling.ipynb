```python
#네이버스토어 크롤링
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains 
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import json
import time


driver=webdriver.Chrome()
driver.get("https://search.shopping.naver.com/ns/search?query=노트북&prevQuery=노트북")

time.sleep(5)

result=[]
result2=[]
items = driver.find_elements(By.CSS_SELECTOR, 'div.basicProductCard_basic_product_card__TdrHT')
#print(items)

time.sleep(3)

for item in items[:5]:
    if item.find_element(By.CSS_SELECTOR, 'strong.basicProductCardInformation_title__Bc_Ng'):
        title = item.find_element(By.CSS_SELECTOR, 'strong.basicProductCardInformation_title__Bc_Ng').text
        price = item.find_element(By.CSS_SELECTOR, 'span.priceTag_inner_price__TctbK').text

        data = item.find_element(By.CSS_SELECTOR, 'a.basicProductCard_link__urzND').get_attribute("data-shp-contents-dtl")
        tmps = json.loads(data)
        for tmp in tmps:
            if tmp['key']=="prod_nm":
                 k1 = tmp['value']
            if tmp['key']=="price":
                 k2 = tmp['value']

        result.append({
            '상품': title,
            '가격': price
        })

        result2.append({
            '상품': k1,
            '가격': k2
        })


print(result)
print()
print(result2)
```

```python
#잡코리아에서 검색 크롤링
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains 
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import json
import time

driver = webdriver.Chrome()
driver.get("https://www.jobkorea.co.kr/")

time.sleep(3)

#검색
search_box=driver.find_element(By.CSS_SELECTOR,"#stext")
search_box.send_keys("파이썬 개발")
search_box.submit()

result=[]
items=driver.find_elements(By.CSS_SELECTOR,'article.list-item')
for item in items[:20]:
    name =item.find_element(By.CSS_SELECTOR, '.corp-name-link').text
    content=item.find_element(By.CSS_SELECTOR,'.information-title-link').text 
    details = item.find_elements(By.CSS_SELECTOR, '.chip-information-group .chip-information-item')
    details_list=[detail.text for detail in details] 
    result.append({
        '이름': name,
        '내용': content,
        '세부사항':details_list
    })

print(json.dumps(result,ensure_ascii=False,indent=3))
```

```python
#제품별 분리
from openpyxl import Workbook, load_workbook
import win32com.client
import os

# 기존 total.xlsx 파일 열기
wb = load_workbook('엑셀자동화/total.xlsx')
ws = wb.active

# 새로운 시트 추가
ws_mouse_pad = wb.create_sheet("마우스 패드")
ws_monitor_32 = wb.create_sheet("32인치 모니터")
ws_keyboard_mechanical = wb.create_sheet("기계식 키보드")
ws_mouse_gaming = wb.create_sheet("게이밍 마우스")

# 각 시트에 데이터를 분배할 목록 (예: 제품명 기준으로 분류)
product_sheets = {
    "마우스 패드": ws_mouse_pad,
    "32인치 모니터": ws_monitor_32,
    "기계식 키보드": ws_keyboard_mechanical,
    "게이밍 마우스": ws_mouse_gaming
}

# 헤더 추가 (순번 없이 제품명, 가격, 수량, 합계)
for sheet in product_sheets.values():
    sheet.append(['제품명', '가격', '수량', '합계'])

# ws에서 데이터를 읽어서 해당 제품 시트에 추가
for row in ws.iter_rows(min_row=2, values_only=True):
    product_name = row[1]  # 제품명 (두 번째 열)
    
    # 제품명에 맞는 시트를 찾아서 데이터 추가
    if product_name in product_sheets:
        sheet = product_sheets[product_name]
        sheet.append([row[1], row[2], row[3], row[4]])  # 제품명, 가격, 수량, 합계

# 변경 사항 저장
wb.save("엑셀자동화/total.xlsx")
wb.close()
```

```python
from openpyxl import Workbook, load_workbook
from openpyxl.styles import Font, Side, Border, Alignment

file_path = '엑셀자동화/total1.xlsx'  
save_path = '엑셀자동화/total_last.xlsx'

wb = load_workbook(file_path)
ws = wb['data']

ws.column_dimensions['B'].width = 20  # 문자수 기준
# ws.column_dimensions['B'].hidden = True  # 열 숨기기
# ws.column_dimensions['B'].bestFit = True  # 열 너비 자동조절
ws.row_dimensions[1].height = 30  # 포인트 기준

# 폰트 변경
ft = Font(size=12, color='0077ff', bold=True, name='돋움')
ws['A1'].font = ft

# 선 속성
border_type1 = Side(color='000000', border_style='thin')
border_type2 = Side(color='000000', border_style='thick')

# 선 지정
border_bottom = Border(bottom=border_type1)  # left, right, top, bottom
ws['A4'].border = border_bottom

# 정렬 지정
alignment_type = Alignment(horizontal='center', vertical='center') 
# horizontal(수평정렬): left, center, right, justify
# vertical(수직정렬): top, center, bottom

ws['B5'].alignment = alignment_type

# 첫 번째 행에 정렬, 선, 폰트 적용
for row in ws.iter_rows(max_row=1):  # 첫 번째 행만 반복
    for cell in row:
        cell.alignment = alignment_type
        cell.border = border_bottom
        cell.font = ft

wb.save(save_path)
wb.close()
```
