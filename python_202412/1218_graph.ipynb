### Matplotlib

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager

#폰트경로설정
fontpath='NanumGothic.ttf'
font_manager.fontManager.addfont(fontpath)
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
```

### .plot()

```python
plt.plot([10,20,30,40])  #y축으로 표시(x축: 1,2,3,,,자동)
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,30,40])  #y축으로 표시(x축: 1,2,3,,,자동)
plt.show()
```

```python
import numpy as np
x=np.arange(1,5) #1,2,3,4
y=np.arange(10,50,10) #10,20,30,40
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([2,4,12,14])
x1=np.array([3,4,5,6])
y1=np.array([5,7,15,17])
plt.plot(x,y,'^',label='세모',markersize=7)
plt.plot(x1,y1,'o',label='동그라미',markersize=5)
plt.legend()  #범례표시
plt.show()
```

```python
x=np.array([1,2,3,4])
y1=np.array([2,4,12,14])
y2=np.array([3,5,13,15])

plt.title('그래프 제목',fontsize=15,color="purple")
plt.plot(x,y1, label='first line',color="pink")
plt.plot(x,y2, label='second line')
plt.legend(title="범례 타이틀",loc="upper right")
plt.show()
```

```python
plt.plot([1,2,3,4],[28,25,26,24])

#기본 그리드 선 추가
plt.grid(axis='both',color='g',linestyle='-.',alpha=0.5,linewidth=0.7)

#보조 그리드 선 추가
plt.minorticks_on()
plt.grid(which='minor',color='gray',linestyle=':',alpha=0.5,linewidth=0.7)

plt.show()
```

```python
x=[1,2,3,4]
y=[78,75,76,71]
plt.plot(x,y)

font={'family':'NanumGothic',
      'color':'b',
      'size':15,
      'weight':'bold'}

plt.xlabel('주간',labelpad=15,fontsize=14,rotation=45)
plt.ylabel('체중',fontdict=font)
plt.show()
```

```python
#범위 직접 지정
x=[1,2,3,4]
y=[28,25,26,21]

plt.plot(x,y)
plt.xlim(0,5)
plt.ylim(10,30)
plt.axis([0,5,10,30])  #[xmin, xmax, ymin, ymax]
plt.show()
```

```python
#눈금설정
plt.plot([78,75,76,72,87,80,77])
# plt.xticks([0,1,2,3],labels=['1달','2달','3달','4달'])
plt.xticks([0,1,2,3,4,5,6],labels=['2024-01-01','2024-01-02','2024-01-03','2024-01-04','2024-01-05','2024-01-06','2024-01-06'],rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
data={
    '성명':['홍길동','김철수','이영희','박민수','정하나'],
    '급여':[3000000,4500000,2500000,5000000,3500000]
}
plt.figure(figsize=(10,5))
plt.plot(data['성명'],data['급여'],marker='o',linestyle='-', color='b')
plt.title('선수별 급여')
plt.xlabel('성명')
plt.ylabel('급여(원)')
plt.grid()
plt.show()
```

### .bar()

```python
x=[1,2,3]
values=[100,700,500]
plt.bar(x,values,color='pink',width=0.8, edgecolor='black')
plt.xticks(x,['2021년','2022년','2023년'])
plt.show()
```

```python
x=[1,2,3,4]
revenue=[20,21,25,25]
profit=[15,18,19,20]

plt.bar(x, revenue, color='green', width=0.6, align='center', label='매출')
plt.bar(x, profit, color='orange', width=0.4, align='edge', label='수익')

plt.xticks(x, ['2021년','2022년','2023년','2024년'])
plt.legend()
plt.show()
```
