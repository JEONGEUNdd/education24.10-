```python
# 12) 성인여부 데이터 변경
import numpy as np
df['성인여부'] = np.where((df['성인여부'] == "woman") | (df['성인여부'] == "man"), "adult", "child")
# df['성인여부'] =df['성인여부'].str.replace('woman','adult').str.replace('man','adult')
# df['성인여부']=df['성인여부'].replace(['man','woman'],'adult')
df
```

```python
# 16) 중복 컬럼 삭제
df=df.drop(['생존여부','객실등급'],axis=1)
df
```

```python
# 17) 탑승지 결측치 행 찾기
df[df['탑승지코드'].isna() & df['탑승지명'].isna()]
```

```python
# 18) 탑승지 결측치 행 삭제
df=df.dropna(subset=['탑승지코드'])  #특정 열기준 결측정보 이용해 전체행 삭제
df
```

```python
# 19) 59~63번 행 출력
df.loc[59:63]
# df.iloc[59:64]
```

```python
# 20) child의 나이 결측치 확인
df[df['나이'].isna()][df['성인여부']=='child']
```

```python
df[df['나이']<0]
```

```python
df['나이']=df['나이'].abs()   #양수로 바꿈
```

```python
# 21) 나이 평균 출력
df['나이'].mean()
# np.mean(df['나이'])
```

```python
# 22) 나이 결측치를 평균값으로 대체
# avg=df['나이'].mean()
# df['나이']=df['나이'].fillna(avg)
df['나이']=df['나이'].fillna(np.mean(df['나이']))
df
```

```python
# 23) 선실코드 종류 출력
df['선실코드'].unique()
```

```python
#선실코드 결측치를 최빈값으로 대체
# df['선실코드'].mode()[0]  #최빈값
df['선실코드']=df['선실코드'].fillna(df['선실코드'].mode()[0])
```
