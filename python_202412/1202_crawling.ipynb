```python
import sqlite3
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# 조회수 증가 함수
def increment_view_count(title):
    conn = sqlite3.connect('view_counts.db')
    cursor = conn.cursor()
    
    # 테이블이 없으면 생성
    cursor.execute('''CREATE TABLE IF NOT EXISTS views (title TEXT PRIMARY KEY, count INTEGER)''')

    cursor.execute('SELECT count FROM views WHERE title = ?', (title,))
    result = cursor.fetchone()
    
    if result:
        new_count = result[0] + 1
        cursor.execute('UPDATE views SET count = ? WHERE title = ?', (new_count, title))
    else:
        new_count = 1  # 새로운 뉴스 제목의 경우 조회수 1로 설정
        cursor.execute('INSERT INTO views (title, count) VALUES (?, ?)', (title, new_count))
    
    conn.commit()
    conn.close()
    
    return new_count

# 셀레니움 웹 드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 네이버 뉴스 페이지 접속
url = 'https://news.naver.com/'
driver.get(url)

# 페이지 로딩 대기
time.sleep(3)

# 뉴스 기사 제목과 링크 추출
articles = driver.find_elements(By.CSS_SELECTOR, 'a.cnf_news._cds_link._editn_link')

# HTML 내용 작성
html_content = """
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <title>네이버 뉴스 기사 목록</title>
    <script type="text/javascript">
        function openPopup(url) {
            window.open(url, 'popup', 'width=800,height=600,scrollbars=yes');
        }
    </script>
</head>
<body>
    <h1>네이버 뉴스</h1>
    <ul>
"""

for article in articles:
    title = article.text  # 뉴스 제목
    link = article.get_attribute('href')  # 링크
    count = increment_view_count(title)  # 조회수 업데이트
    html_content += f'<li><a href="javascript:void(0);" onclick="openPopup(\'{link}\')">{title} (조회수: {count}회)</a></li>'

html_content += """
    </ul>
</body>
</html>
"""

# HTML 파일로 저장
with open("news_list1.html", "w", encoding="utf-8") as file:
    file.write(html_content)

# 웹 드라이버 종료
driver.quit()

print("뉴스 목록이 'news_list1.html' 파일로 저장되었습니다.")
```
