```python
#students.txt파일을 딕셔너리로 불러오기
def read_students(filename):
    students={}
    with open(filename, "r", encoding="utf-8") as f:
        header=f.readline().strip().split(',')

        #헤더를 키로 초기화
        for key in header:
            students[key]=[]
            
        for line in f:
            values=line.strip().split(',')
            for i in range(len(header)):
                students[header[i]].append(values[i])
    return students
            
filename_students='students.txt'
students_dict=read_students(filename_students)
print(students_dict)
```

```python
#계산기 프로그램
def show_cal():
    print("\n===계산기===")
    print("1. 덧셈")
    print("2. 뺄셈")
    print("3. 곱셈")
    print("4. 나눗셈")
    print("5. 종료")

def add(num1,num2):
    print(f"결과: {num1 + num2}")
def substract(num1,num2):
    print(f"결과: {num1 + num2}")
def multiply(num1,num2):
    print(f"결과: {num1 + num2}")
def divide(num1,num2):
    print(f"결과: {num1 + num2}")
    
def main():
    while True:
        show_cal()
        choice=input("연산을 선택하세요:")
        if choice=="5":
            print("프로그램 종료")
            break
        num1=float(input("첫 번째 숫자: "))
        num2=float(input("두 번째 숫자: "))

        if choice=="1":
            add(num1,num2)
        elif choice=="2":
            substract(num1,num2)
        elif choice=="3":
            multiply(num1,num2)
        elif choice=="4":
            divide(num1,num2)
main()  
```

```python
#디렉토리(target_dir) 내 파일들을 확장자별 폴더로 분류
# 확장자를 기준으로 폴더를 생성한 뒤, 해당 폴더로 파일을 이동

def organize_files(target_dir):   #디렉토리에서 파일 목록을 읽고, 확장자별로 분류
    for filename in os.listdir(target_dir):  #target_dir 내의 모든 파일과 폴더를 리스트로 반환
        source_file=os.path.join(target_dir,filename) #filename과 target_dir를 결합해 파일의 절대 경로 생성
        if(os.path.isfile(source_file)):
            ext=filename.split('.')[-1]  #리스트의 마지막 요소를 가져와 확장자를 추출
            ext_dir=os.path.join(target_dir,ext)  #target_dir와 확장자 이름(ext)을 결합해 확장자 폴더 경로 생성.
            if not os.path.exists(ext_dir):    #확장자 폴더가 없으면 생성
                os.makedirs(ext_dir)
                print(f"확장자가 생성되었습니다. - {ext}")
            new_file=os.path.join(ext_dir,filename)
            os.rename(source_file,new_file)
            print(f"{filename}파일이 {ext_dir}로 옮겨졌습니다.")
            
target_directory = os.path.join(os.getcwd(),"sample")
organize_files(target_directory)
```

```python
class expense():
    def __init__(self):
        self.expenses=[]
      
    def show_menu(self):
        print("====가계부====")
        print("1. 지출 입력")
        print("2. 지출 목록 보기")
        print("3. 총 지출 확인")
        print("4. 종료")

    def add(self):
        date=input("날짜를 입력하세요 (예: 11-19):")
        item=input("지출 항목을 입력하세요")
        amount=int(input("금액을 입력하세요"))
    
        self.expenses.append([date,item,amount])  #리스트
        print("지출이 추가되었습니다")
        
    def show_list(self):
        print("\n날짜\t항목\t금액")
        print("-"*20)
        for date, item, amount in self.expenses:
            print(f"{date}\t{item}\t{amount:,}원")
    
    def show_total(self):
        total = 0 
        for date, item, amount in self.expenses:
            total+=amount
        print(f"\n총 지출: {total:,}원")
       
    def run(self):
        while True:
            self.show_menu()
            choice=input("원하는 작업을 선택하세요")
            if choice=="1":
                self.add()
            elif choice=="2":
                self.show_list()
            elif choice=="3":
                self.show_total()
            elif choice=="4":
                print("프로그램을 종료합니다.")
                break
a=expense()
a.run() 
```

```python
#네이버뉴스에서 키워드 검색해 기사제목, 링크 추출
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
from datetime import datetime

def get_news_titles(keyword,num_pages):
    driver=webdriver.Chrome()
    titles=[]
    
    for page in range(1,num_pages+1):
        url=f"https://search.naver.com/search.naver?where=news&query={keyword}&start={(page-1)*10+1}"
        driver.get(url)
        time.sleep(2)
        
        soup=BeautifulSoup(driver.page_source,'html.parser')
        
        main_titles=soup.select("a.news_tit")

        for title in main_titles:
            titles.append({
                'title':title.get('title'),
                'link':title.get('href')
            })
        print(f"{page}페이지 완료")

    print("\n===검색 결과===")
    print(f"검색어: {keyword}")
    print(f"총 {len(titles)}개의 기사 수집\n")

    for idx,item in enumerate(titles,1):
        print(f"{idx}.{item['title']}")
        print(f"   링크: {item['link']}")
        print()
    driver.quit()

keyword="빅데이터"
num_pages=2
get_news_titles(keyword,num_pages)
```

```python
#format_sample.xlsx불러와서 스타일적용해 새로운파일로 만들기
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill, Alignment

def format_worksheet(filename): 
    wb = load_workbook(filename)
    ws = wb.active

    header_font=Font(bold=True, size=12)
    header_fill=PatternFill(start_color="DDDDDD",end_color="DDDDDD",fill_type="solid")

    for cell in ws[1]:
        cell.font=header_font
        cell.fill=header_fill
        cell.alignment=Alignment(horizontal='center')

    for row in ws.iter_rows(min_row=2):
        for cell in row:
            cell.alignment=Alignment(horizontal='left')

    wb.save('formatted_file.xlsx')
    
format_worksheet('format_sample.xlsx')
```

```python
#뉴스 기사 크롤링해서 워드 클라우드
# https://www.flaticon.com/icon-fonts-most-downloaded?weight=solid&type=uicon    #아이콘 사이트
import requests
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
from wordcloud import WordCloud
from konlpy.tag import Okt
from PIL import Image

url="https://n.news.naver.com/mnews/hotissue/article/003/0012959761?type=series&cid=1088290"
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')

title=soup.select_one("#title_area").text.strip()
content=soup.select_one("#dic_area").text.strip()

def get_compound_nouns(text):

    okt = Okt()
    tokens = text.split()
    combined_nouns = []
    current_combined = ""

    for token in tokens:
        nouns = okt.nouns(token)

        for noun in nouns:
            if current_combined:
                current_combined += noun
            else:
                current_combined = noun

            if not (current_combined and (nouns.index(noun) < len(nouns) - 1)):
                combined_nouns.append(current_combined)
                current_combined = ""

    if current_combined:
        combined_nouns.append(current_combined)

    return combined_nouns

#한글 세팅(워드클라우드, 맵플로립)
fontpath=r"C:\Users\enjoy\python_202412\fonts\DoHyeon-Regular.ttf"
font_manager.fontManager.addfont(fontpath)
plt.rc('font',family='Do Hyeon')
plt.rcParams['axes.unicode_minus']=False

words=get_compound_nouns(content)
# print(words)

unique_words, counts=np.unique(words, return_counts=True)
freq=dict(zip(unique_words, counts))

# print(freq)

mask_image=np.array(Image.open('heart.jpg'))

wordcloud = WordCloud(
    width=800, 
    height=400,
    background_color='white',
    font_path=fontpath,  # Windows의 한글 폰트 경로
    mask=mask_image,
    contour_color='white',  #경계선 색
    contour_width=1,
    colormap='prism'
).generate_from_frequencies(freq)

# WordCloud 시각화
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('단어 빈도 워드클라우드')
# plt.saveImg('워드클라우드.png',bbox_inches='tight')  #이미지로 저장
plt.show()
```
