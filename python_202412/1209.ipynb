```python
##finditer():정규식과 매치되는 모든 문자열을 반복 가능한 객체로 리턴

string="abc123def456"
pattern=r"\d+"
matches=re.finditer(pattern,string)
for match in matches:
  print(f"{match.group()} 시작 인덱스:{match.start()}, 끝 인덱스:{match.end()}")
```

```python
import re
pattern=re.compile(r"\d+")
string="abc123def456"
result=pattern.search(string)

if result:
  print("일치: ",result.group(0))  #문자열에서 숫자를 검색하고, 일치 항목을 출력하거나, 문자열 내의 모든 숫자를 리스트로 반환
else:
  print("불일치")

result=pattern.findall(string)
print(result)

for match in pattern.finditer(string):
  print(f"Found email: {match.group()}")
  print(f"Start position: {match.start()}")
  print(f"End position: {match.end()}")
  print(f"Start, End position: {match.span()}")
  print("-"*20)
```

```python
import re

text="""
<html>
<head><title>Sample Page</title></head>
<body>
<h1>Welcome to My Page</h1>
<p>This is a paragraph.</p>
<p>Another paragraph with <a href="https://example.com">a link</a>.</p>
<p>Another paragraph with <a href="https://example2.com">a link</a>.</p>
<ul>
<li>Item 1</li>
<li>Item 2</li>
<li>Item 3</li>
</ul>
<p>Contact: user@example.com</p>
</body>
</html>
"""

#제목 태그 추출
title_pattern=r"<title>(.*?)</title>"
title_match=re.search(title_pattern,text)
if title_match:
  title=title_match.group(1)
  print("제목:", title)

#모든 문단내용 추출
paragraph_pattern=r"<p>(.*?)</p>"
paragraphs=re.findall(paragraph_pattern,text,re.DOTALL | re.IGNORECASE)
#re.DOTALL:.가 줄바꿈문자를 포함한 모든 문자와 매칭되도록함
#IGNORECASE:대소문자 구분하지 않고 매칭
print("모든 문단내용:")
for paragraph in paragraphs:
  print(paragraph)

#모든 리스트 아이템 추출
pattern3=r"<li>(.*?)</li>"
items=re.findall(pattern3,text)
print("리스트 아이템:",items)

#이메일주소 추출
email_pattern=r"[\w]+@[\w]+.[\w]+"
emails=re.findall(email_pattern,text)
print("이메일 주소:",emails)

#모든 링크 추출
link_pattern=r'href="([^"]+)"'
link_pattern=r'href="(.*?)"'
links=re.findall(link_pattern,text)
print("모든 링크:",links)
```

```python
#기본적으로 정규식은 가능한 많은 문자 매칭하려함(greedy)
#비탐욕적(non-greedy) 문자 또는 비탐욕적 패턴은 정규식에서 가능한 한 적은 문자 매칭

'''비탐욕적
.*?
.+?
\d+?
'''

import re
text="<div>Hello</div><div>World</div>"

greedy_pattern=r"<div>(.*)</div>"
greedy_result=re.findall(greedy_pattern,text)
print("Greedy Result:", greedy_result)

non_greedy_pattern=r"<div>(.*?)</div>"
non_greedy_result=re.findall(non_greedy_pattern,text)
print("Non-Greedy Result:", non_greedy_result)
```

```python
import re

data = """
park 800903-1049123
kim 780412-2034889
"""

# 첫 번째 정규식
pat = re.compile(r"(\d{6})[-]\d{7}")
result = pat.sub(r"\1-******", data)
print(result)

# 두 번째 정규식
pat1 = r'(\d{6}-)(\d)\d{6}'
result1 = re.sub(pat1, r'\1\2******', data)
print(result1)
```

```python
import requests
import re
from bs4 import BeautifulSoup

url="http://lecture.pul.kr/ex6.html"
response=requests.get(url)

# print(response.text)

# HTML 내용
html_content = response.text

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html_content, 'html.parser')

#이메일 주소 추출
email_pattern = r"[\w\.-]+@[\w\.-]+\.\w+"
emails = re.findall(email_pattern, html_content)
print("이메일 주소:", emails)

#전화번호 추출
#(형식: (123) 456-7890 또는 987-654-3210)
phone_pattern = r"\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}"
# 정규식을 사용하여 전화번호 추출
phone_numbers = re.findall(phone_pattern, html_content)
print("전화번호:", phone_numbers)

#가격 추출
price_pattern = r"\$\d+\.\d{2}"
prices = re.findall(price_pattern, html_content)
print("가격:", prices)

#링크 추출
link_pattern = r'href="([^"]+)"'
links = re.findall(link_pattern, html_content)
print("링크:", links)


#특수문자 추출
special_char_pattern = r"[^a-zA-Z0-9\s]"
special_characters = re.findall(special_char_pattern, html_content)
print("특수 문자:", special_characters)

#주석 제거
comment_pattern = r"<!--.*?-->"
cleaned_html = re.sub(comment_pattern, "", html_content)
print("주석 제거된 HTML:")
print(cleaned_html)


#괄호 안의 내용 추출
parentheses_pattern = r"\((.*?)\)"
parentheses_contents = re.findall(parentheses_pattern, html_content)
print("괄호 안의 내용:", parentheses_contents)


#모든 텍스트 추출
all_text = soup.get_text()
print("모든 텍스트 추출:",all_text)


#제품 이름 추출
product_names = [product.h3.get_text() for product in soup.find_all('div', class_='product')]
print("제품 이름:")
for name in product_names:
    print(name)

# 제품 가격 추출
prices = [price.get_text() for price in soup.find_all('p', string=lambda text: text and text.startswith('Price:'))]
print("\n제품 가격:")
for price in prices:
    print(price)

# Footer 내용 추출
footer_content = soup.find('footer').get_text(strip=True)
print("\nFooter 내용:")
print(footer_content)


#모든 링크의 텍스트 추출(링크 텍스트)
link_texts = [link.get_text() for link in soup.find_all('a')]
print("모든 링크 텍스트:")
for text in link_texts:
    print(text)

# 모든 li 요소의 내용 추출
list_items = [li.get_text() for li in soup.find_all('li')]
print("\n모든 li 요소의 내용:")
for item in list_items:
    print(item)

# 특정 단어 찾기 (예: "Price")
price_keywords = [text for text in soup.stripped_strings if 'Price' in text]
print("\n특정 단어(Price) 찾기:")
for keyword in price_keywords:
    print(keyword)

# 특정 CSS 클래스의 요소 추출 (class: 'product')
products = [product.get_text() for product in soup.find_all(class_='product')]
print("\n특정 CSS 클래스('product')의 요소 추출:")
for product in products:
    print(product)


# 모든 헤더(헤더 태그: h1, h2, h3 등) 추출
headers = [header.get_text() for header in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])]
print("모든 헤더 텍스트:")
for header in headers:
    print(header)

#  p 요소 안의 텍스트 추출 (단락)
paragraphs = [p.get_text() for p in soup.find_all('p')]
print("\n모든 p 요소 안의 텍스트:")
for paragraph in paragraphs:
    print(paragraph)

#  특정 키워드가 포함된 li 요소 찾기 (키워드: Banana)
keyword = "Banana"
keyword_items = [li.get_text() for li in soup.find_all('li') if keyword in li.get_text()]
print(f"\n키워드 '{keyword}'가 포함된 li 요소:")
for item in keyword_items:
    print(item)

#  특정 가격 이상인 제품 찾기 (가격 20.00 이상)
products_above_20 = []
for product in soup.find_all('div', class_='product'):
    price = product.find('p', string=lambda text: text and text.startswith('Price:'))
    if price:
        price_value = float(price.get_text().split('$')[1])
        if price_value >= 20.00:
            products_above_20.append(product.h3.get_text())  # 제품 이름
print("\n가격 20.00 이상인 제품:")
for product in products_above_20:
    print(product)

#  HTML에서 특정 텍스트를 대체하기 (텍스트: Sample HTML Document)
new_html = soup.prettify().replace("Sample HTML Document", "Updated HTML Document")
print("\nHTML에서 텍스트 대체:")
print(new_html)

```
