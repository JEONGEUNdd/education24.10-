```python
import pandas as pd
df = pd.DataFrame({
    '제품': ['노트북', '태블릿', '스마트폰', '노트북', '스마트폰', '태블릿'],
    '지역': ['서울', '부산', '서울', '대구', '부산', '서울'],
    '판매량': [10, 5, 8, 12, 6, 4],
    '가격': [1200000, 500000, 800000, 1300000, 750000, 550000],
    '판매자번호': ['A01', 'A02', 'A01', 'A03', 'A02', 'A03']
})
df
```

```python
df.groupby('제품')[['판매량']].mean()
```

```python
df.groupby(['지역','제품'])[['판매량']].sum()
```

```python
df['매출액']=df['판매량']*df['가격']
df
```

```python
df.groupby('판매자번호')[['매출액']].sum()
```

```python
#집계함수
df.groupby('제품').agg(
    {
        '판매량':['sum','mean','count'],
        '가격':['min','max']
    }
)
```

```python
df = pd.DataFrame({
    '제품': ['A', 'A', 'B', 'B', 'C'],
    '가격': [1000, 1200, 800, 850, 900],
    '판매량': [10, 15, 20, 25, 30],
    '평점': [4.5, 4.0, 4.8, 4.2, 4.6]
})
df
```

```python
df.groupby('제품').mean(numeric_only=True)   #계산할 수 있는 항목만 처리  
```

```python
df.groupby('제품').agg({
    '가격':'mean',
    '판매량':'sum',
    '평점':'max'
})
```

```python
df.groupby('제품').agg({
    '가격':['mean','min','max'],
    '판매량':['sum','count'],
    '평점':['mean','std']
})
```

```python
def range_calc(x):
    return x.max()-x.min()
    
df.groupby('제품').agg({
    '가격':['mean',range_calc],
    '판매량':'sum'
})
```
